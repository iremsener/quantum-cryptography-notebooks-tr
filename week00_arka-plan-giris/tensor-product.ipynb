## 0.3 Tensör Çarpımı: Kübitler Nasıl Birleştirilir? (Tensor Product: how to combine qubits)

İki farklı kübitimiz ($A$ ve $B$) olduğunu hayal edelim. $A$ sisteminin durumunu $|\psi\rangle_A$, $B$ sisteminin durumunu ise $|\phi\rangle_B$ olarak tanımlayabileceğimizi biliyoruz. Peki $A$ ve $B$ sistemlerinin birleşik durumunu ($|\psi\rangle_{AB}$) birlikte nasıl yazabiliriz?

Bu ortak durumu (joint state) hesaplamanın kuralı, **tensör çarpımı** (bazen **Kronecker çarpımı** olarak da adlandırılır) ile verilir. İki kübit için durumları şu şekilde yazarsak:

$$|\psi\rangle_A = \alpha_A |0\rangle_A + \beta_A |1\rangle_A = \begin{pmatrix} \alpha_A \\ \beta_A \end{pmatrix}$$

$$|\phi\rangle_B = \alpha_B |0\rangle_B + \beta_B |1\rangle_B = \begin{pmatrix} \alpha_B \\ \beta_B \end{pmatrix}$$

Ortak durum $|\psi\rangle_{AB} \in \mathbb{C}^2 \otimes \mathbb{C}^2$, her bir bağımsız $|\psi\rangle_A$ ve $|\phi\rangle_B$ vektörünün tensör çarpımı olarak ifade edilebilir:

$$|\psi\rangle_{AB} = |\psi\rangle_A \otimes |\phi\rangle_B = \begin{pmatrix} \alpha_A \\ \beta_A \end{pmatrix} \otimes |\phi\rangle_B = \begin{pmatrix} \alpha_A |\phi\rangle_B \\ \beta_A |\phi\rangle_B \end{pmatrix} = \begin{pmatrix} \alpha_A \alpha_B \\ \alpha_A \beta_B \\ \beta_A \alpha_B \\ \beta_A \beta_B \end{pmatrix}$$


Tahmin edebileceğiniz gibi, eğer $A$ ve $B$ kuantum sistemleri tek bir kübit'ten daha büyükse, bu sistemlerin durumlarını da elbette aynı şekilde birleştirebiliriz. İki genel vektörün tensör çarpımının genel tanımı şu şekildedir:

> 🟦 **Tanım 0.3.1 — Tensör Çarpımı.** Sırasıyla $|\psi_1\rangle \in \mathbb{C}^{d_1}$ ve $|\psi_2\rangle \in \mathbb{C}^{d_2}$ olmak üzere iki vektör verildiğinde, tensör çarpımı şu şekilde hesaplanır:
> 
> $$|\psi_1\rangle \otimes |\psi_2\rangle = \begin{pmatrix} \alpha_1 \\ \vdots \\ \alpha_d \end{pmatrix} \otimes |\psi_2\rangle = \begin{pmatrix} \alpha_1 |\psi_2\rangle \\ \vdots \\ \alpha_d |\psi_2\rangle \end{pmatrix}$$
> 
> ve elde edilen $|\psi_1\rangle \otimes |\psi_2\rangle$ bileşke vektörü $\mathbb{C}^{d_1} \otimes \mathbb{C}^{d_2}$ durum uzayında yer alır.

*(Not: EPR kısaltması Einstein, Podolsky ve Rosen isimlerini temsil eder. İlerleyen bölümlerde bu durumun dolanık/entangled olduğunu göstereceğim.)*


```python
import numpy as np

# |0> ve |1> durum vektörleri
psi_A = np.array([[1], [0]]) # |0> durumu
phi_B = np.array([[0], [1]]) # |1> durumu

# Tensör (Kronecker) çarpımı ile |01> durumunu elde etme
psi_AB = np.kron(psi_A, phi_B)
print("Bileşke Vektör |01>:\n", psi_AB)
```

Kuantum bilgi teorisinde aşağıdaki basitleştirilmiş (veya pratik) gösterimler yaygın olarak kullanılır:

*   **Tensör çarpım sembolünün yazılmaması:** 
    $$|\psi\rangle_A \otimes |\Psi\rangle_B = |\psi\rangle_A |\Psi\rangle_B$$
*   **Klasik bitlerin yan yana bir dizge olarak yazılması:** 
    $$|0\rangle_A \otimes |0\rangle_B = |0\rangle_A |0\rangle_B = |00\rangle_{AB}$$
*   **Aynı durumdaki birden fazla sistemin birleştirilmesi (Tensör Gücü):** 
    $$|\psi\rangle_1 \otimes |\psi\rangle_2 \dots \otimes |\psi\rangle_n = |\psi\rangle^{\otimes n}$$


> 📜 **Önerme 0.3.1** Tensör çarpımı işlemi birkaç kullanışlı matematiksel özelliği sağlar:
> 
> 1. **Dağılma Özelliği (Distributive):** 
>    $$|\psi_1\rangle \otimes (|\psi_2\rangle + |\psi_3\rangle) = |\psi_1\rangle \otimes |\psi_2\rangle + |\psi_1\rangle \otimes |\psi_3\rangle$$
>    Benzer şekilde:
>    $$(|\psi_1\rangle + |\psi_2\rangle) \otimes |\psi_3\rangle = |\psi_1\rangle \otimes |\psi_3\rangle + |\psi_2\rangle \otimes |\psi_3\rangle$$
> 
> 2. **Birleşme Özelliği (Associative):** 
>    $$|\psi_1\rangle \otimes (|\psi_2\rangle \otimes |\psi_3\rangle) = (|\psi_1\rangle \otimes |\psi_2\rangle) \otimes |\psi_3\rangle$$
> 
> 3. **Değişme Özelliği Yoktur (NOT commutative):** Genel olarak, eğer $|\psi_1\rangle = |\psi_2\rangle$ durumu hariç tutulursa:
>    $$|\psi_1\rangle \otimes |\psi_2\rangle \neq |\psi_2\rangle \otimes |\psi_1\rangle$$

Bu ilişkiler sadece ket vektörleri ($| \rangle$) için değil, aynı zamanda bra vektörleri ($\langle |$) için de geçerlidir.

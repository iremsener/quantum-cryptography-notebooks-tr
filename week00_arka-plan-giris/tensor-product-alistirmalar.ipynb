### Tensor Product Alıştırmalar

> 📝 **Örnek 0.3.1** Tensör çarpımı kuralını kullanarak iki tekil kübitin standart tabanından, iki kübitlik bileşke standart tabanı yeniden elde edelim. İki kübitlik bir $AB$ sistemi için standart tabanın şu şekilde verildiğini hatırlayalım:
> 
> $$|00\rangle_{AB} = \begin{pmatrix} 1 \\ 0 \\ 0 \\ 0 \end{pmatrix}, \quad |01\rangle_{AB} = \begin{pmatrix} 0 \\ 1 \\ 0 \\ 0 \end{pmatrix}, \quad |10\rangle_{AB} = \begin{pmatrix} 0 \\ 0 \\ 1 \\ 0 \end{pmatrix}, \quad |11\rangle_{AB} = \begin{pmatrix} 0 \\ 0 \\ 0 \\ 1 \end{pmatrix}$$


Bu taban, her bir tekil kübit için standart taban elemanlarının tensör çarpımı alınarak inşa edilebilir: $|0\rangle_A \otimes |0\rangle_B$, $|0\rangle_A \otimes |1\rangle_B$, $|1\rangle_A \otimes |0\rangle_B$, $|1\rangle_A \otimes |1\rangle_B$. 

Örnek olarak, $|10\rangle_{AB}$ durumunun nasıl elde edildiğini açıkça inceleyelim:

$$|1\rangle_A \otimes |0\rangle_B = \begin{pmatrix} 0 \\ 1 \end{pmatrix} \otimes |0\rangle_B = \begin{pmatrix} 0 \cdot |0\rangle_B \\ 1 \cdot |0\rangle_B \end{pmatrix} = \begin{pmatrix} 0 \cdot 1 \\ 0 \cdot 0 \\ 1 \cdot 1 \\ 1 \cdot 0 \end{pmatrix} = \begin{pmatrix} 0 \\ 0 \\ 1 \\ 0 \end{pmatrix} = |10\rangle_{AB}$$

Daha önce iki kübitlik durumlara ait birkaç başka örnek de görmüştük. Şimdi tensör çarpımını kullanarak bunları iki bağımsız tekil kübit durumundan elde edip edemeyeceğimizi inceleyelim.


> 📝 **Örnek 0.3.2** $|+\rangle_A = \frac{1}{\sqrt{2}}(|0\rangle_A + |1\rangle_A)$ ve $|1\rangle_B$ durumlarını ele alalım. Bu iki sistemin bileşke durumu olan $|\psi\rangle_{AB}$, tensör çarpımı ile şu şekilde verilir:
> 
> $$|\psi\rangle_{AB} = |+\rangle_A \otimes |1\rangle_B = \frac{1}{\sqrt{2}} \begin{pmatrix} 1 \\ 1 \end{pmatrix} \otimes |1\rangle_B = \frac{1}{\sqrt{2}} \begin{pmatrix} 1 \cdot |1\rangle_B \\ 1 \cdot |1\rangle_B \end{pmatrix} = \frac{1}{\sqrt{2}} \begin{pmatrix} 0 \\ 1 \\ 0 \\ 1 \end{pmatrix}$$


Bileşke durumu, tensör çarpımının dağılma özelliğini kullanarak standart taban cinsinden cebirsel olarak da şu şekilde ifade edebiliriz:

$$|\psi\rangle_{AB} = \frac{1}{\sqrt{2}}(|0\rangle_A + |1\rangle_A) \otimes |1\rangle_B$$

$$= \frac{1}{\sqrt{2}}(|0\rangle_A \otimes |1\rangle_B + |1\rangle_A \otimes |1\rangle_B)$$

$$= \frac{1}{\sqrt{2}}(|01\rangle_{AB} + |11\rangle_{AB})$$


> 📝 **Örnek 0.3.3** $|+\rangle_A = \frac{1}{\sqrt{2}}(|0\rangle_A + |1\rangle_A)$ ve $|+\rangle_B = \frac{1}{\sqrt{2}}(|0\rangle_B + |1\rangle_B)$ durumlarını ele alalım. Bu iki sistemin bileşke durumu olan $|\psi\rangle_{AB}$ şu şekilde hesaplanır:
> 
> $$|\psi\rangle_{AB} = \frac{1}{\sqrt{2}}(|0\rangle_A + |1\rangle_A) \otimes \frac{1}{\sqrt{2}}(|0\rangle_B + |1\rangle_B)$$
> 
> $$= \frac{1}{2}(|00\rangle_{AB} + |01\rangle_{AB} + |10\rangle_{AB} + |11\rangle_{AB})$$
> 
> $$= \frac{1}{2} \begin{pmatrix} 1 \\ 1 \\ 1 \\ 1 \end{pmatrix}$$
> 



> 📝 **Örnek 0.3.4** Aşağıdaki kuantum durumunu ele alalım:
> 
> $$|\Psi\rangle_{AB} = \frac{1}{\sqrt{2}}(|+\rangle_A |+\rangle_B + |-\rangle_A |-\rangle_B)$$
> 
> Terimleri tek tek açarak bu durumu standart (hesapsal) taban cinsinden ifade edelim. Önceki örneklerden ve tanımlardan yararlanarak çarpımları hesaplayalım:
> 
> $$|+\rangle_A |+\rangle_B = \frac{1}{2}(|0\rangle_A + |1\rangle_A)(|0\rangle_B + |1\rangle_B) = \frac{1}{2}(|00\rangle_{AB} + |01\rangle_{AB} + |10\rangle_{AB} + |11\rangle_{AB})$$
> 
> $$|-\rangle_A |-\rangle_B = \frac{1}{2}(|0\rangle_A - |1\rangle_A)(|0\rangle_B - |1\rangle_B) = \frac{1}{2}(|00\rangle_{AB} - |01\rangle_{AB} - |10\rangle_{AB} + |11\rangle_{AB})$$




$$|\Psi\rangle_{AB} = \frac{1}{\sqrt{2}}(|+\rangle_A |+\rangle_B + |-\rangle_A |-\rangle_B)$$

$$= \frac{1}{2\sqrt{2}}(|00\rangle_{AB} + |01\rangle_{AB} + |10\rangle_{AB} + |11\rangle_{AB} + |00\rangle_{AB} - |01\rangle_{AB} - |10\rangle_{AB} + |11\rangle_{AB})$$

Dikkat edilirse zıt işaretli terimler ($|01\rangle$ ve $|10\rangle$) birbirini götürür. Geriye kalan terimleri sadeleştirdiğimizde şu sonucu elde ederiz:

$$= \frac{1}{\sqrt{2}}(|00\rangle_{AB} + |11\rangle_{AB}) = |\text{EPR}\rangle_{AB}$$



```python
import numpy as np

# Temel Tekli Qubit Durumları
qubit_0 = np.array([1, 0])
qubit_1 = np.array([0, 1])

plus = (qubit_0 + qubit_1) / np.sqrt(2)
minus = (qubit_0 - qubit_1) / np.sqrt(2)

# 1. Yol: Hadamard Tabanındaki Terimlerin Tensör Çarpımı ve Toplamı
plus_plus = np.kron(plus, plus)
minus_minus = np.kron(minus, minus)
durum_hadamard = (plus_plus + minus_minus) / np.sqrt(2)

# 2. Yol: Standart Taban Cinsinden EPR Çifti
zero_zero = np.kron(qubit_0, qubit_0)
one_one = np.kron(qubit_1, qubit_1)
durum_standart = (zero_zero + one_one) / np.sqrt(2)

# İki durumun birbirine eşit olup olmadığının kontrolü
print("Hadamard tabanından gelen matris:\n", durum_hadamard)
print("\nStandart tabandan gelen matris:\n", durum_standart)
print("\nİki gösterim eşit mi?:", np.allclose(durum_hadamard, durum_standart))
```

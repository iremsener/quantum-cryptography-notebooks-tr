## 0.5 Kübitler Üzerindeki Dönüşümler (Transformations on qubits)

Tıpkı klasik bitlerde olduğu gibi, kübitler üzerinde de işlemler gerçekleştirebiliriz. Kuantum durumlarını vektörler olarak yazabildiğimize göre, vektörleri yine vektörlere eşleyen bir $U$ doğrusal operatörü (lineer operatör) arıyoruz:

$$|\psi_{\text{out}}\rangle = U |\psi_{\text{in}}\rangle$$

Buradaki bazı $U$ matrisleri için; eğer $|\psi_{\text{in}}\rangle \in \mathbb{C}^d$ ise, $U$ karmaşık elemanlara sahip $d \times d$ boyutunda bir matristir. Herhangi bir kuantum durumu için $\langle \psi | \psi \rangle = 1$ olduğunu hatırlayalım. Bunun oldukça önemli olduğunu görmüştük; çünkü durumu ölçtüğümüzde olasılıkların toplamının da 1 olması gerektiğini bize söylüyordu. Bu durum, $U$ işleminin **iç çarpımı koruması gerektiği** anlamına gelir, yani:

$$\langle \psi_{\text{out}} | \psi_{\text{out}} \rangle = \langle \psi_{\text{in}} | U^\dagger U | \psi_{\text{in}} \rangle = 1$$

Benzer şekilde, aynı durum $U^\dagger$ işlemi için de geçerli olmalıdır:

$$\langle \psi_{\text{out}} | \psi_{\text{out}} \rangle = \langle \psi_{\text{in}} | U U^\dagger | \psi_{\text{in}} \rangle = 1$$

> **Not:** $(U|\psi\rangle)^\dagger = \langle \psi|U^\dagger$ eşitliğini unutmayınız.


Olasılıkları korumak adına, $U$ işleminin herhangi bir vektörün uzunluğunu (normunu) koruması gerektiğini görüyoruz. Bu durum, tam olarak $\mathbb{I}$ birim matris olmak üzere $U^\dagger U = U U^\dagger = \mathbb{I}$ şartı sağlandığında gerçekleşir. Bu tür bir $\mathbb{I}$ matrisi bu notlar boyunca sürekli karşımıza çıkacaktır ve onu aşağıda tanımlayalım:

> 🟦 **Tanım 0.5.1 — Birim Matris (Identity).** Birim matris $\mathbb{I}$, köşegen üzerindeki her bir elemanı 1'e eşit olan köşegen, kare bir matristir, yani:
> 
> $$\mathbb{I} = \begin{pmatrix} 1 & 0 & \dots & \dots & 0 \\ 0 & 1 & \dots & \dots & 0 \\ \vdots & \vdots & \ddots & \vdots & \vdots \\ 0 & 0 & \dots & 0 & 1 \end{pmatrix}$$


Herhangi bir $d$ boyutu için, $d \times d$ boyutundaki birim matrisi $\mathbb{I}_d$ olarak gösteririz.

> 🌐 **Kural:** Birim matris, tüm kuantum durumlarını koruyan uniter bir işlemdir. Yani herhangi bir $|\psi\rangle$ kuantum durumu için, $\mathbb{I}|\psi\rangle = |\psi\rangle$ eşitliği geçerlidir.

Bağlamdan anlaşılabildiği durumlarda birim matrisin boyutunu açıkça belirtmeyeceğiz. Kuantum rejiminde izin verilen yegane işlemler **uniter işlemlerdir (unitary operations)**.

> 🟦 **Tanım 0.5.2 — Uniter İşlem (Unitary Operation).** Bir $U$ işleminin uniter olabilmesi için gerek ve yeter şart $U^\dagger U = U U^\dagger = \mathbb{I}$ olmasıdır.




> 📝 **Örnek 0.5.1** Aşağıdaki matrisi ele alalım:
> 
> $$H = \frac{1}{\sqrt{2}} \begin{pmatrix} 1 & 1 \\ 1 & -1 \end{pmatrix}$$
> 
> Bu matris için $H^\dagger = H$ olduğunu ve dolayısıyla şu eşitliğin sağlandığını kendiniz de görebilirsiniz:
> 
> $$H^\dagger H = H H = \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix} = \mathbb{I}$$
> 
> Yani, $H$ matrisi uniterdir. Bu matrisin $|0\rangle$ durumu üzerindeki etkisi şöyledir:
> 
> $$H|0\rangle = \frac{1}{\sqrt{2}} \begin{pmatrix} 1 & 1 \\ 1 & -1 \end{pmatrix} \begin{pmatrix} 1 \\ 0 \end{pmatrix} = \frac{1}{\sqrt{2}} \begin{pmatrix} 1 \\ 1 \end{pmatrix} = |+\rangle$$

Benzer şekilde, $H|1\rangle = |-\rangle$ olduğunu da kendiniz doğrulayabilirsiniz. Böylece $H$ işleminin, hesapsal tabanı ($\{|0\rangle, |1\rangle\}$) Hadamard tabanına ($\{|+\rangle, |-\rangle\}$) dönüştürdüğünü görmüş oluruz. Tam da bu nedenden dolayı $H$ işlemine **Hadamard Dönüşümü (Hadamard Transform / Hadamard Kapısı)** adı verilir.

Birim matrisin ($\mathbb{I}$) kendisinin de hiçbir şeyi değiştirmeyen (durumu olduğu gibi bırakan) uniter bir işlem olduğunu not düşelim. Şimdi biraz daha karmaşık bir işlemi ele alalım.


> 📝 **Örnek 0.5.2** Herhangi bir $\theta \in \mathbb{R}$ için aşağıdaki matrisi ele alalım:
> 
> $$R(\theta) = \begin{pmatrix} \cos \frac{\theta}{2} & -\sin \frac{\theta}{2} \\ \sin \frac{\theta}{2} & \cos \frac{\theta}{2} \end{pmatrix}$$
> 
> Bu matrisin eşlenik transpozu (adjoint) şu şekildedir:
> 
> $$R^\dagger(\theta) = \begin{pmatrix} \cos \frac{\theta}{2} & \sin \frac{\theta}{2} \\ -\sin \frac{\theta}{2} & \cos \frac{\theta}{2} \end{pmatrix}$$
> 
> Ve dolayısıyla matrislerin çarpımı bize şu sonucu verir:
> 
> $$R(\theta)R^\dagger(\theta) = \begin{pmatrix} \cos \frac{\theta}{2} & -\sin \frac{\theta}{2} \\ \sin \frac{\theta}{2} & \cos \frac{\theta}{2} \end{pmatrix} \cdot \begin{pmatrix} \cos \frac{\theta}{2} & \sin \frac{\theta}{2} \\ -\sin \frac{\theta}{2} & \cos \frac{\theta}{2} \end{pmatrix}$$
> 
> $$= \begin{pmatrix} \cos^2 \frac{\theta}{2} + \sin^2 \frac{\theta}{2} & 0 \\ 0 & \sin^2 \frac{\theta}{2} + \cos^2 \frac{\theta}{2} \end{pmatrix} = \begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix} = \mathbb{I}$$
> 
> Benzer şekilde $R^\dagger(\theta)R(\theta) = \mathbb{I}$ olduğu da kontrol edilebilir, bu yüzden $R(\theta)$ matrisi uniterdir.


Bu dönme işleminin $|0\rangle$ ve $|1\rangle$ hesapsal taban vektörleri üzerindeki etkisi aşağıdaki gibidir:

$$R(\theta)|0\rangle = \begin{pmatrix} \cos \frac{\theta}{2} & -\sin \frac{\theta}{2} \\ \sin \frac{\theta}{2} & \cos \frac{\theta}{2} \end{pmatrix} \begin{pmatrix} 1 \\ 0 \end{pmatrix} = \begin{pmatrix} \cos \frac{\theta}{2} \\ \sin \frac{\theta}{2} \end{pmatrix}$$

$$R(\theta)|1\rangle = \begin{pmatrix} \cos \frac{\theta}{2} & -\sin \frac{\theta}{2} \\ \sin \frac{\theta}{2} & \cos \frac{\theta}{2} \end{pmatrix} \begin{pmatrix} 0 \\ 1 \end{pmatrix} = \begin{pmatrix} -\sin \frac{\theta}{2} \\ \cos \frac{\theta}{2} \end{pmatrix}$$

Eğer özel bir durum olarak $\theta = \frac{\pi}{2}$ alırsak; $\cos \frac{\theta}{2} = \sin \frac{\theta}{2} = \cos \frac{\pi}{4} = \frac{1}{\sqrt{2}}$ olur ve böylece şu sonuçları elde ederiz:

$$R\left(\frac{\pi}{2}\right)|0\rangle = |+\rangle \quad \text{ve} \quad R\left(\frac{\pi}{2}\right)|1\rangle = -|-\rangle$$


### 0.5.1 Uniter İşlemler Olarak Pauli Matrisleri (Pauli matrices as unitary operations)

Bu bölümde, genellikle $X, Y, Z$ sembolleri ile gösterilen **Pauli matrislerine** göz atacağız. Fizikte oldukça ünlü olan bu matrisler, ileride göreceğimiz üzere kuantum hesaplamada **bit çevirme (bit flip)** ve **evre çevirme (phase flip)** işlemleri olarak son derece ilginç ve önemli yorumlara sahiptir. Pauli matrisleri, aşağıdaki formlara sahip uniter $2 \times 2$ matrislerdir:

$$X = \begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix}$$

$$Z = \begin{pmatrix} 1 & 0 \\ 0 & -1 \end{pmatrix}$$

$$Y = iXZ = \begin{pmatrix} 0 & -i \\ i & 0 \end{pmatrix}$$

Pauli-X matrisi, standart taban vektörlerinin yerlerini birbirleriyle değiştirerek (yer değiştirme) etki eder:

$$X |0\rangle = |1\rangle$$

$$X |1\rangle = |0\rangle$$


Klasik hesaplamadaki karşılığına benzer şekilde, 0'ı 1'e ve 1'i 0'ı çevirdiği için $X$ matrisine **NOT (Değil) kapısı** da denir. Bu işlem literatürde **bit çevirme (bit flip)** işlemi olarak da bilinir. 

Öte yandan, Pauli-Z matrisi standart taban üzerinde bir **evre çevirme (phase flip)** gerçekleştirerek etki eder:

$$Z |0\rangle = |0\rangle$$

$$Z |1\rangle = -|1\rangle$$

Pauli-Z matrisinin Hadamard tabanındaki etkisi ise $|+\rangle$ ve $|-\rangle$ vektörlerinin yerlerini birbiriyle değiştirmektir. Açıkça yazmak gerekirse:

$$Z |+\rangle = Z(|0\rangle + |1\rangle)/\sqrt{2} = (Z|0\rangle + Z|1\rangle)/\sqrt{2} = (|0\rangle - |1\rangle)/\sqrt{2} = |-\rangle$$


Benzer şekilde, $Z|-\rangle = |+\rangle$ olduğu görülür. Böylece $Z$ matrisinin standart tabanda bir evre çevirme (phase flip) gibi davranırken, Hadamard tabanı üzerinde ise bir bit çevirme (bit flip) gibi davrandığını görmüş oluruz. 

Sisteme hem bir bit çevirme hem de bir evre çevirme işlemini birlikte uygulamak bize $Y = iXZ$ matrisini verir. Buradaki $i$ çarpanı $Y$ matrisini **Hermityen (Hermitian)** yapar, yani $Y^\dagger = Y$ eşitliği sağlanır. Bu matris, standart taban vektörlerine uygulandığında hem bir bit çevirme hem de bir evre çevirme etkisi yaratır:

$$Y |0\rangle = iXZ |0\rangle = iX |0\rangle = i|1\rangle$$

$$Y |1\rangle = -iXZ |0\rangle = -iX |1\rangle = -i|0\rangle$$




```python
import numpy as np

# Temel qubit durumları
q0 = np.array([1, 0]) # |0>
q1 = np.array([0, 1]) # |1>

# Pauli Matrisleri
X = np.array([[0, 1], [1, 0]])
Z = np.array([[1, 0], [0, -1]])

# X kapısının |0> üzerindeki bit-flip etkisi: X|0> = |1>
print("X |0> Sonucu:", np.dot(X, q0)) 

# Z kapısının |1> üzerindeki phase-flip etkisi: Z|1> = -|1>
print("Z |1> Sonucu:", np.dot(Z, q1))
```

## 0.6 Kopyalanamama İlkesi! (No cloning!)

Bu bölümde, klasik bitlerin aksine, rastgele seçilmiş qubitlerin kopyalanamayacağını göstereceğiz. Burada, derste gösterilenden biraz daha farklı bir ispat sunuyoruz. 

Eğer elimizde kopyalama yapabilen uniter bir $C$ işlemi olsaydı, bu işlem herhangi bir girdi $|\psi\rangle$ kübit durumu için bize $C(|\psi\rangle \otimes |0\rangle) = |\psi\rangle \otimes |\psi\rangle$ sonucunu vermeliydi. Çelişki yoluyla ispat (olmayana ergi) yöntemini kullanmak adına, böyle uniter bir işlemin var olduğunu varsayalım. Bu durumda, böyle bir uniter işlem bize özel olarak şu iki eşitliği vermelidir:

$$C(|\psi_1\rangle \otimes |0\rangle) = |\psi_1\rangle \otimes |\psi_1\rangle$$

$$C(|\psi_2\rangle \otimes |0\rangle) = |\psi_2\rangle \otimes |\psi_2\rangle$$


$C$ bir uniter işlem olduğundan, $C^\dagger C = \mathbb{I}$ eşitliğine sahibiz ve bu sayede iç çarpım üzerinden şu adımları yazabiliriz:

$$\langle\psi_1 | \psi_2\rangle = \langle\psi_1 | \psi_2\rangle \langle0|0\rangle$$

$$= \Big( (\langle\psi_1| \otimes \langle0|) C^\dagger C (|\psi_2\rangle \otimes |0\rangle) \Big)$$

$$= \Big( (\langle\psi_1| \otimes \langle\psi_1|) (|\psi_2\rangle \otimes |\psi_2\rangle) \Big) = (\langle\psi_1 | \psi_2\rangle)^2$$

Açıkça görüldüğü üzere, $0 < |\langle\psi_1 | \psi_2\rangle| < 1$ olduğu her durumda yukarıdaki eşitlik ($\langle\psi_1 | \psi_2\rangle = (\langle\psi_1 | \psi_2\rangle)^2$) geçerliliğini yitirir. Dolayısıyla, böyle bir kopyalama uniter operatörü $C$ **asla var olamaz**. Örneğin, $|\psi_1\rangle = |0\rangle$ ve $|\psi_2\rangle = |+\rangle$ durumları tam olarak bu koşulu (birbirlerine ne tamamen dik ne de tamamen paralel olma durumunu) sağlar.


Rastgele kuantum durumlarını kopyalayamıyor oluşumuz, kuantum bilgisinin klasik bilgiden gerçekten ne kadar farklı olduğunu gösterir. Bu durum ayrıca daha derin bir anlayış kazanmamızı sağlar: Tek bir $|\Psi\rangle = \alpha |0\rangle + \beta |1\rangle$ kübit kopyasından $\alpha$ ve $\beta$ değerlerini asla tam olarak belirleyemeyeceğimiz anlamına gelir. Aksini düşündüğümüzde; eğer belirleyebilseydik, $\alpha |0\rangle + \beta |1\rangle$ süperpozisyonunu hazırlayan bir makine yaparak o kübiti kolayca kopyalayabilirdik.

Bu durumun bazı güzel özellikleri olsa da (örneğin doğası gereği gelen bir **kopya koruma mekanizması** sağlaması gibi), aynı zamanda kübitlerin çok değerli ve hassas olduğu anlamına gelir. Örneğin bir qubit göndermeye çalışırken başarısız olduğumuzda, klasik bitlerde yaptığımız gibi "basitçe yeniden göndermeyi" deneyemeyiz. Eğer videolarda beni net bir şekilde duyamasaydınız, videoyu geri sarabilir, sesi açabilir ve tekrar deneyebilirdiniz. Ancak size kübitler üzerinden konuşuyor olsaydım, bunu yapmanın **hiçbir yolu olmayacaktı!**

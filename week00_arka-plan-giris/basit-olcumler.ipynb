## 0.4 Basit Ölçümler (Simple measurements)

Bir kübit'i ölçtüğümüzde ne olduğunu inceleyelim. Klasik olarak, bir bitin ölçülmesini basitçe bir okuma işlemi olarak düşünebilirsiniz: elimizde '0' ve '1' durumunu kodlayan bir sistem vardır ve hangisi olduğunu bulmak için bir ölçüm yaparız.

### 0.4.1 Standart Tabanda Ölçüm (Measurement in the standard basis)

Öncelikle tek bir kübit'i ele alalım. Kuantum ölçümleri olasılıksal sonuçlar doğurabilir; bu da kuantum bilgisi ile klasik bilgi arasındaki temel ve köklü farkı vurgular. 

Örneğin, eğer $|\psi\rangle \in \mathbb{C}^2$ durumu $|0\rangle$ ve $|1\rangle$ arasında bir süperpozisyonsa, $|\psi\rangle$ durumu üzerinde ölçüm yaptığımızda, belirli bir olasılık dağılımına karşılık gelen farklı ölçüm sonuçları elde ederiz. Peki bu olasılıklar nasıl üretilir? 

Farklı sonuçların olasılığı (örneğin '0' sonucu için), kabaca söylersek, "kübit vektörümüzün içinde aslında ne kadar '0' olduğuna bakılarak" hesaplanabilir. Bu durum, $|\psi\rangle$ ile $|0\rangle$ arasındaki **iç çarpım (inner product)** ile nicelendirilir. 

Daha somut olarak, $\alpha, \beta$ karmaşık sayılar olmak üzere $|\psi\rangle = \alpha |0\rangle + \beta |1\rangle$ şeklindeki tek bir kübit durumunu ele alalım. Bu kübit ölçüldüğünde, $p_0$ olasılığıyla "0" sonucu ve $p_1$ olasılığıyla "1" sonucu elde edilir. Bu olasılıklar, iç çarpımların hesaplanmasıyla belirlenebilir:

$$p_0 = |\langle \psi | 0 \rangle|^2 = \left| \begin{pmatrix} \alpha^* & \beta^* \end{pmatrix} \begin{pmatrix} 1 \\ 0 \end{pmatrix} \right|^2 = |\alpha|^2$$

$$p_1 = |\langle \psi | 1 \rangle|^2 = \left| \begin{pmatrix} \alpha^* & \beta^* \end{pmatrix} \begin{pmatrix} 0 \\ 1 \end{pmatrix} \right|^2 = |\beta|^2$$


Şimdi, önceki kısımlarında bahsettiğimiz $|\alpha|^2 + |\beta|^2 = 1$ koşulunun arkasındaki gerçek ve mantıklı nedeni görebiliyoruz: Bu koşul, $p_0 + p_1 = 1$ anlamına gelir; yani '0' ve '1' sonuçlarını gözlemleme olasılıklarının toplamı **tam olarak 1'e (kesinliğe)** eşittir. 

Kuantum bilgisayar biliminde, sonuçları $|0\rangle$ için '0' ve $|1\rangle$ için '1' olarak etiketlemek adettendir. Öte yandan, fizikçiler genellikle $|0\rangle$ için $+1$ ve $|1\rangle$ için $-1$ etiketlerini kullanmayı tercih ederler.


### Uygulama: Deterministik bir süreçten rastlantısallık elde etmek (Application: Randomness from a deterministic process)

Şimdiye kadar öğrendiklerimizle ilginç bir şey yapabilir miyiz? Cevabın evet olduğu ortaya çıkıyor: Sadece tekil kübitler hazırlayıp bunları standart tabanda ölçerek, ilkesel olarak klasik yöntemlerle yapılması imkansız olan bir görevi başarabiliriz. Yani, **gerçek rastgele sayılar (true random numbers)** üretebiliriz. 

Şekil 1'de gösterilen şu süreci ele alalım: İlk olarak, bir kübit'i $|+\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)$ durumunda hazırlayın. Ardından, bu durumu standart tabanda ölçün. Her bir sonucu elde etme olasılığı, iç çarpımların hesaplanmasıyla şu şekilde bulunabilir:

$$p_0 = |\langle + | 0 \rangle|^2 = \left| \frac{1}{\sqrt{2}} (\langle 0 | + \langle 1 |) | 0 \rangle \right|^2 = \left| \frac{1}{\sqrt{2}} (\underbrace{\langle 0 | 0 \rangle}_{1} + \underbrace{\langle 1 | 0 \rangle}_{0}) \right|^2 = \frac{1}{(\sqrt{2})^2} = \frac{1}{2}$$

$$p_1 = |\langle + | 1 \rangle|^2 = \left| \frac{1}{\sqrt{2}} (\langle 0 | + \langle 1 |) | 1 \rangle \right|^2 = \left| \frac{1}{\sqrt{2}} (\underbrace{\langle 0 | 1 \rangle}_{0} + \underbrace{\langle 1 | 1 \rangle}_{1}) \right|^2 = \frac{1}{(\sqrt{2})^2} = \frac{1}{2}$$

> **Şekil 1:** Süperpozisyondaki bir kübit'in hazırlanması ve ölçülmesiyle gerçek rastlantısallığın (genuine randomness) üretilmesi.


Bu basit örnek bize kuantum bilgisinin gücü hakkında çoktan bir şeyler söylüyor: Belirli bir kurala göre (deterministik olarak) $|+\rangle$ qubitini hazırlayan ve ardından standart tabanda ölçüm yapan bir makine inşa edebiliriz. $p_0 = p_1 = 1/2$ olduğundan, bu makine bize kusursuz bir rastgele sayı üretme imkanı tanır; üstelik makinemizin içinde **hiçbir rastlantısal süreç kullanılmamış olmasına rağmen!** Buna karşılık, hiçbir klasik deterministik makinenin sıfırdan gerçek rastgele sayılar üretemeyeceği ispatlanabilir.

Tek bir kübit'in standart tabanda nasıl ölçüleceğini gördük. Ölçüm sonuçlarının olasılıklarını hesaplama kuralı, doğrudan $n$-qubitlik durumların ölçülmesine de genelleştirilebilir. Gerçekten de, şu şekildeki bir $n$-qubit kuantum durumunu ele alalım:

$$|\Psi\rangle = \sum_{x \in \{0,1\}^n} \alpha_x |x\rangle$$

$|\Psi\rangle$ durumu $\{|x\rangle\}_x$ standart tabanında ölçüldüğünde ne olur? Herhangi bir $x$ sonucunun gelme olasılığının $p_x = |\langle x | \Psi \rangle|^2 = |\alpha_x|^2$ ile verildiği görülür. Bu durum, $|\Psi\rangle$ vektörünün normalizasyon şartına olan ihtiyacını bir kez daha açıklar.


```python
import numpy as np

# Şekil 1'deki deterministik |+> durumunu hazırlayalım (Süperpozisyon)
qubit_0 = np.array([1, 0])
qubit_1 = np.array([0, 1])
plus_state = (qubit_0 + qubit_1) / np.sqrt(2)

# Olasılıklar: p0 = |<0|+>|^2 ve p1 = |<1|+>|^2
p0 = np.abs(np.dot(qubit_0, plus_state))**2
p1 = np.abs(np.dot(qubit_1, plus_state))**2

print(f"Teorik Olasılıklar -> p0 (0 gelme): {p0:.1f}, p1 (1 gelme): {p1:.1f}")

# Simülasyon: Bu kuantum makinesini 10 kez çalıştırıp rastgele bitler üretelim
sonuclar = np.random.choice([0, 1], size=10, p=[p0, p1])
print("Üretilen Gerçek Rastgele Bit Dizisi:", "".join(map(str, sonuclar)))
```

### 0.4.2 Bir kübit'i farklı tabanlarda ölçmek (Measuring a qubit in other bases)

Kübitimizi başka bir tabanda ölçebilir miyiz? Bu sorunun cevabı evet! Aslında bu durum, kuantum bilgisini, geçerli olan tek tabanın standart taban olduğu klasik bilgiden ayıran başka bir temel özelliktir. 

Böyle daha genel bir durumu nasıl analiz edeceğimizi bulmak için öncelikle bir adım geri atalım ve yukarıdaki olasılıkları nasıl bulduğumuzu hatırlayalım. Standart tabanda ölçüm yaparken olasılıklar, durum standart taban cinsinden yazıldığında genliklerin karesiyle veriliyordu. 

$\mathcal{G} = \{|v\rangle, |v^\perp\rangle\}$ vektörleri tarafından verilen farklı bir ortonormal tabanda bir kübit ölçümü yaparken, kübit durumunu bu yeni taban cinsinden ifade etmemiz gerektiği sezgisel olarak anlaşılabilir. Yani, aşağıdaki eşitliği sağlayacak şekilde yeni $\hat{\alpha}$ ve $\hat{\beta}$ genliklerini bulmamız gerekir:

$$|\psi\rangle = \alpha |0\rangle + \beta |1\rangle = \hat{\alpha} |v\rangle + \hat{\beta} |v^\perp\rangle$$

> **Not:** Burada $x$ genel olarak "$|x\rangle$" sonuçları için etiket olarak kullanılır. $|v^\perp\rangle$ (v-dik) ifadesi ise $|v\rangle$ vektörüne dik olan taban vektörünü temsil eder.


> 📝 **Örnek 0.4.1** Bir örnek olarak, yine $|+\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle)$ qubit durumunu ele alalım. Bu durumu standart tabanda ölçmek yerine, şimdi $|+\rangle$ ve $|-\rangle = \frac{1}{\sqrt{2}}(|0\rangle - |1\rangle)$ şeklindeki iki ortonormal vektör tarafından verilen $\mathcal{H} = \{|+\rangle, |-\rangle\}$ (Hadamard) tabanında ölçelim. 
> 
> Açıkça görülebileceği üzere, bu kübit durumunu yeni taban cinsinden şu şekilde yazabiliriz:
> 
> $$|+\rangle = 1 \cdot |+\rangle + 0 \cdot |-\rangle$$
> 
> Dolayısıyla, bu ölçüm sonucunda "$|+\rangle$" durumunu elde etme olasılığımız tam olarak $1$ (yani %100) olur. Buradan yola çıkarak, kuantum ölçüm sonuçlarının olasılıklarının, **ölçüm yaptığımız tabana (baz) dramatik bir şekilde bağlı olduğunu** net bir şekilde görebiliriz.


> 📝 **Örnek 0.4.2** Rastgele bir $\alpha|0\rangle + \beta|1\rangle$ kübit durumunun $\{|+\rangle, |-\rangle\}$ tabanında ölçülmesini ele alalım. Bu kübiti yeni taban cinsinden nasıl ifade edeceğimizi bulmak için, öncelikle $|0\rangle$ ve $|1\rangle$ taban elemanlarının bu yeni tabanda nasıl göründüğünü belirlemek kolaylık sağlar. Buradan şu eşitlikleri elde ederiz:
> 
> $$|0\rangle = \frac{1}{2}\left[(|0\rangle + |1\rangle) + (|0\rangle - |1\rangle)\right] = \frac{1}{\sqrt{2}}(|+\rangle + |-\rangle)$$
> 
> $$|1\rangle = \frac{1}{2}\left[(|0\rangle + |1\rangle) - (|0\rangle - |1\rangle)\right] = \frac{1}{\sqrt{2}}(|+\rangle - |-\rangle)$$
> 
> Böylece rastgele durumumuzu şu şekilde yazabiliriz:
> 
> $$\alpha|0\rangle + \beta|1\rangle = \frac{1}{\sqrt{2}}\left[\alpha(|+\rangle + |-\rangle) + \beta(|+\rangle - |-\rangle)\right]$$
> 
> $$= \frac{\alpha + \beta}{\sqrt{2}}|+\rangle + \frac{\alpha - \beta}{\sqrt{2}}|-\rangle$$
> 
> Bu dönüşüm, ölçüm sonucunda $\frac{|\alpha + \beta|^2}{2}$ olasılıkla "$|+\rangle$" sonucunu ve $\frac{|\alpha - \beta|^2}{2}$ olasılıkla "$|-\rangle$" sonucunu elde edeceğimiz anlamına gelir.


> 🟦 **Alıştırma 0.4.1** $|\Psi\rangle = |0\rangle$ durumunu ele alalım. 
> 1. $|\Psi\rangle$ durumunu **standart tabanda** ölçtüğümüzde elde edeceğimiz $p_0$ ve $p_1$ olasılıkları nelerdir?
> 2. $|\Psi\rangle$ durumunu **Hadamard tabanında** ölçtüğümüzde elde edeceğimiz $p_+$ ve $p_-$ olasılıkları nelerdir?

*(Cevap İpucu: Standart tabanda ölçülürse $p_0=1, p_1=0$ olur. Hadamard tabanında ölçülürse $p_+=1/2, p_-=1/2$ olur.)*


Çoğu zaman tüm olasılık dağılımını hesaplamakla ilgilenmeyiz; sadece belirli bir sonucun gelme olasılığını bilmek isteriz. Tüm $|\psi\rangle$ durumunu yeni bir tabanda en baştan yazmak yerine, bu spesifik olasılığı bulmanın daha verimli bir yolu var mıdır?

Bunu araştırmak için yine tek bir kübit durumunu ele alalım:

$$|\psi\rangle = \alpha |0\rangle + \beta |1\rangle$$

Standart taban elemanlarının birbiriyle ortonormal olduğunu hatırlayalım. Matematiksel olarak bu durum şu anlama gelir (burada $\dagger$ sembolü eşlenik transpozu temsil eder):

$$(|0\rangle)^\dagger |0\rangle = \begin{pmatrix} 1 & 0 \end{pmatrix} \begin{pmatrix} 1 \\ 0 \end{pmatrix} = 1$$

$$(|0\rangle)^\dagger |1\rangle = \begin{pmatrix} 1 & 0 \end{pmatrix} \begin{pmatrix} 0 \\ 1 \end{pmatrix} = 0$$

Vektörler ortonormal olduğu için, yukarıda iddia edildiği gibi, hedeflediğimiz olasılıkları sadece iki vektör arasındaki **iç çarpımı (inner product)** hesaplayarak doğrudan bulabiliriz. 


Özellikle, $|\psi\rangle = \alpha |0\rangle + \beta |1\rangle$ kübit durumu verildiğinde, "$|0\rangle$" ve "$|1\rangle$" sonuçlarını şu olasılıklarla elde ederiz:

$$p_0 = |\langle 0 | \psi \rangle|^2 = \left| \begin{pmatrix} 1 & 0 \end{pmatrix} \begin{pmatrix} \alpha \\ \beta \end{pmatrix} \right|^2 = |\alpha|^2$$

$$p_1 = |\langle 1 | \psi \rangle|^2 = \left| \begin{pmatrix} 0 & 1 \end{pmatrix} \begin{pmatrix} \alpha \\ \beta \end{pmatrix} \right|^2 = |\beta|^2$$


> 📝 **Örnek 0.4.3** $|0\rangle$ durumunu $\mathcal{H}$ Hadamard tabanında (yukarıya bakın) ölçtüğümüzü varsayalım. "$|+\rangle$" ve "$|-\rangle$" sonuçlarını gözlemleme olasılıkları şu şekilde verilir:
> 
> $$p_+ = |\langle + | 0 \rangle|^2 = \left| \begin{pmatrix} 1/\sqrt{2} & 1/\sqrt{2} \end{pmatrix} \begin{pmatrix} 1 \\ 0 \end{pmatrix} \right|^2 = \frac{1}{2}$$
> 
> $$p_- = |\langle - | 0 \rangle|^2 = \left| \begin{pmatrix} 1/\sqrt{2} & -1/\sqrt{2} \end{pmatrix} \begin{pmatrix} 1 \\ 0 \end{pmatrix} \right|^2 = \frac{1}{2}$$


Çoklu kübitler için olasılıkları bulma kuralı da tamamen benzerdir (analogdur).

> 🟦 **Tanım 0.4.1** Bir $|\psi\rangle$ kuantum durumunu $\{|b_j\rangle\}_{j=1}^d$ ortonormal tabanında ölçtüğümüzü varsayalım. "$b_j$" sonucunu gözlemleme olasılığı şu hesaplama ile bulunabilir:
> 
> $$p_j = |\langle b_j | \psi \rangle|^2$$
> 
> Ölçüm sonucunda "$b_j$" sonucu elde edildiğinde, **ölçüm sonrası durum (post-measurement state)** $|b_j\rangle$ vektörü ile verilir.

Farklı tabanlarda kuantum sistemlerini ölçmeye dair sezgi kazanmak için şimdi bazı örneklere göz atalım. İlk olarak tek bir qubit örneğini inceleyelim.


> 📝 **Örnek 0.4.4** $|\Psi\rangle = \frac{1}{\sqrt{2}}(|0\rangle + i|1\rangle)$ kübit durumunu ele alalım ve bu kübiti $\{|+\rangle, |-\rangle\}$ tabanında ölçelim. "$+$" ve "$-$" sonuçlarını elde etme olasılıkları aşağıdaki gibi değerlendirilebilir:
> 
> $$p_+ = |\langle + | \Psi \rangle|^2 = \left| \frac{1}{2} (\langle 0 | + \langle 1 |)(|0\rangle + i|1\rangle) \right|^2$$
> 
> $$= \frac{1}{4} \left| \langle 0|0\rangle + \langle 0|1\rangle i + \langle 1|0\rangle + \langle 1|1\rangle i \right|^2$$
> 
> $$= \frac{1}{4} |1 + i|^2$$
> 
> $$= \frac{1}{4}(1 - i)(1 + i) = \frac{1}{2}$$
> 
> $$p_- = |\langle - | \Psi \rangle|^2 = \left| \frac{1}{2} (\langle 0 | - \langle 1 |)(|0\rangle + i|1\rangle) \right|^2$$
> 
> $$= \frac{1}{4} \left| \langle 0|0\rangle + \langle 0|1\rangle i - \langle 1|0\rangle - \langle 1|1\rangle i \right|^2$$
> 
> $$= \frac{1}{4} |1 - i|^2$$
> 
> $$= \frac{1}{4}(1 + i)(1 - i) = \frac{1}{2}$$

Bu örnek, işin içine giren durumlar karmaşık değerli (complex-valued) genliklere sahip olduğunda, iç çarpımı değerlendirirken ekstra dikkatli olunması gerektiğini göstermektedir: yani $\langle \Psi |$ bra vektörünü alırken ...işin içinde karmaşık bir sayı olduğunda her zaman +/- işaretini değiştirmeyi (yani eşleniğini almayı) unutmayın (çünkü $\langle \Psi |$ bra vektörü, $|\Psi \rangle$ ket vektörünün **eşlenik transpozudur / conjugate transpose**).

Genel olarak $n$-kübit sistemlerden bahsedecek olsak da, elbette $|0\rangle$, $|1\rangle$ ve $|2\rangle$ olmak üzere üç seviyeden (boyuttan) oluşan bir kuantum sistemini, yani bir **qutrit** sistemini de ele alabiliriz. Bu sistemlerde de ölçüm sonuçlarının olasılıklarını elde etme kuralı tamamen aynı kalır.



```python
import numpy as np

# |Psi> = (|0> + i|1>) / sqrt(2) durumunun ket (sütun) vektörü
# Python'da sanal birim "i" yerine "1j" kullanılır
psi_ket = np.array([1, 1j]) / np.sqrt(2)

# Bra vektörünü bulmak için hem transpozunu alıp hem de eşleniğini (hermitian conjugate) hesaplıyoruz
psi_bra = np.conj(psi_ket) # İşaretin otomatik eksiye döndüğünü kontrol edebilirsiniz

print("Ket Vektörü:\n", psi_ket)
print("\nBra Vektörü (Eşlenik Alınmış):\n", psi_bra)
```

> 📝 **Örnek 0.4.5** Aşağıdaki vektör ile temsil edilen, 3 boyutlu bir kuantum sistemi olan bir **qutrit** durumunu ele alalım:
> 
> $$|v\rangle = \frac{1}{\sqrt{2}} \begin{pmatrix} 1 \\ 0 \\ 0 \end{pmatrix} + \frac{1}{2} \begin{pmatrix} 0 \\ 1 \\ 0 \end{pmatrix} + \frac{1}{2} \begin{pmatrix} 0 \\ 0 \\ 1 \end{pmatrix}$$
> 
> Ve bu durumu, taban vektörleri aşağıdakiler gibi verilen $\mathcal{B} = \{|b_1\rangle, |b_2\rangle, |b_3\rangle\}$ tabanında ölçelim:
> 
> $$|b_1\rangle = \begin{pmatrix} 1 \\ 0 \\ 0 \end{pmatrix}, \quad |b_2\rangle = \frac{1}{\sqrt{2}} \begin{pmatrix} 0 \\ 1 \\ 1 \end{pmatrix}, \quad |b_3\rangle = \frac{1}{\sqrt{2}} \begin{pmatrix} 0 \\ 1 \\ -1 \end{pmatrix}$$
> 
> Her bir ölçüm sonucunu elde etme olasılıkları şu şekilde hesaplanabilir:
> 
> $$p_{b_1} = |\langle b_1 | v \rangle|^2 = \frac{1}{2}$$
> 
> $$p_{b_2} = |\langle b_2 | v \rangle|^2 = \langle b_2 | v \rangle \langle v | b_2 \rangle = \frac{1}{2\sqrt{2}}(1 + 1) \cdot \frac{1}{2\sqrt{2}}(1 + 1) = \frac{1}{2}$$
> 
> $$p_{b_3} = |\langle b_3 | v \rangle|^2 = \langle b_3 | v \rangle \langle v | b_3 \rangle = \frac{1}{2\sqrt{2}}(1 - 1) \cdot \frac{1}{2\sqrt{2}}(1 - 1) = 0$$


### Beklenen Değerler (Expectation values)

Fizikçiler (ve bilgisayar bilimciler de!) ölçüm sonuçlarının **beklenen değerlerini (expectation values)** hesaplamayı çok severler. Çünkü bu değerler, bir ölçümün çok kez tekrarlanması durumunda sistemin göstereceği ortalama davranış hakkında bilgi verir (ancak ileride göreceğimiz üzere, ölçüm yapmak sistemin durumunu bozacaktır/çökertecektir!). 

Bir $|\Psi\rangle$ kübitini $\{|0\rangle, |1\rangle\}$ standart tabanında ölçtüğümüzü varsayalım. Ayrıca, fizik dünyasındaki yaygın kabulü benimseyerek bu sonuçları $\pm 1$ ile etiketleyelim. Bu durum; "$|0\rangle$" sonucunu $+1$ çıktısı ile, "$|1\rangle$" sonucunu ise $-1$ çıktısı ile ilişkilendirdiğimiz anlamına gelir. $|\Psi\rangle$ ölçüldüğünde elde edilen sonucun beklenen değeri ($E$) o halde şu şekilde olur:

$$E = 1 \cdot |\langle 0 | \psi \rangle|^2 - 1 \cdot |\langle 1 | \psi \rangle|^2$$

$|\langle 0 | \psi \rangle|^2 = \langle \psi | 0 \rangle \langle 0 | \psi \rangle$ olduğunu göz önünde bulundurursak, şu eşitliği yazabiliriz:

$$E = \langle \psi | (|0\rangle\langle0| - |1\rangle\langle1|) | \psi \rangle = \langle \psi | Z | \psi \rangle$$

Burada $Z = |0\rangle\langle0| - |1\rangle\langle1|$ olarak tanımlanmıştır. İleride daha detaylı göreceğimiz üzere bu $Z$ ifadesi, kuantum hesaplamanın en temel operatörlerinden biri olan **Pauli-Z matrisidir**.

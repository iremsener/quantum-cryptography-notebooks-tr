## 0.7 Bloch Küresi (Bloch sphere)

Tekil kübitler için, **Bloch küresi** olarak adlandırılan oldukça kullanışlı bir görsel temsil (gösterim) mevcuttur. Bu tür güzel bir geometrik gösterimin **yalnızca tekil kübitler için** geçerli olduğunu not etmek gerekir. Bu gösterimi elde etmek için, kübit durumunu şu şekilde ifade ederiz:

$$|\psi\rangle = e^{i\gamma} \left( \cos \frac{\theta}{2} |0\rangle + e^{i\phi} \sin \frac{\theta}{2} |1\rangle \right)$$

Burada $\gamma$, $\theta$ ve $\phi$ birer reel sayıdır. Dışarıdaki $e^{i\gamma}$ **küresel evresi (global phase)**, ölçüm sonuçlarının olasılıkları üzerinde gözlemlenebilir hiçbir etkiye sahip olmadığı için ihmal edilir.


Küresel evrenin neden önemsiz olduğunu görmek için, bazı reel $\gamma_1, \gamma_2$ sayıları için aşağıdaki iki durumu ele alalım:

$$|\psi_1\rangle = e^{i\gamma_1} \left( \cos \frac{\theta}{2} |0\rangle + e^{i\phi} \sin \frac{\theta}{2} |1\rangle \right)$$

$$|\psi_2\rangle = e^{i\gamma_2} \left( \cos \frac{\theta}{2} |0\rangle + e^{i\phi} \sin \frac{\theta}{2} |1\rangle \right)$$

Dikkat edilirse bu iki durum arasındaki ilişki $|\psi_1\rangle = e^{i(\gamma_1 - \gamma_2)} |\psi_2\rangle$ şeklindedir. Buna göre, herhangi bir $\{|b\rangle\}_b$ tabanına göre yapılacak bir ölçümde, herhangi bir $b$ sonucunu elde etme olasılığı her iki durum için de tamamen birbirine eşittir, çünkü:

$$|\langle \psi_1 | b \rangle|^2 = \langle b | \psi_1 \rangle \langle \psi_1 | b \rangle = e^{i(\gamma_1 - \gamma_2)} e^{-i(\gamma_1 - \gamma_2)} \langle b | \psi_2 \rangle \langle \psi_2 | b \rangle = |\langle \psi_2 | b \rangle|^2$$


Ayrıca, bu parametrizasyonun normalizasyon şartını koruduğunu da unutmamak gerekir; çünkü $|\alpha|^2 + |\beta|^2 = \cos^2(\theta/2) + \sin^2(\theta/2) = 1$ eşitliği her zaman sağlanır. 

Böylece $(\theta, \phi)$ sayıları cinsinden, bir kübit'i Şekil 2'de gösterildiği gibi 3 boyutlu bir küre üzerindeki bir nokta olarak düşünebiliriz. Bu kürenin, $|v\rangle \in \mathbb{C}^2$ vektörleri için kullandığımız koordinat sisteminin aynısını takip etmediği, aksine bizim bu yeni koordinat sistemine (küresel koordinatlara) bir dönüşüm yapmamız gerektiği önemle vurgulanmalıdır.


```python
# Eğer yüklü değilse notebook içinde: !pip install qiskit-aer matplotlib
from qiskit.visualization import plot_bloch_vector
import matplotlib.pyplot as plt

# Notta bahsedilen theta ve phi koordinatlarına göre küre üzerinde bir nokta çizelim
# Örnek olarak |+> durumunu (theta = pi/2, phi = 0) kartezyen koordinatta (1, 0, 0) olarak çiziyoruz
plot_bloch_vector([1, 0, 0], title="Bloch Kuresi Uzerinde |+> Durumu")
plt.show()
```

> 🟦 **Tanım 0.7.1 — Bloch Küresi Gösterimi.** Bir qubit durumunun aşağıdaki parametrizasyonu ($\theta, \phi$):
> 
> $$|\psi\rangle = e^{i\gamma} \left( \cos \frac{\theta}{2} |0\rangle + e^{i\phi} \sin \frac{\theta}{2} |1\rangle \right)$$
> 
> **Bloch küresi gösterimi** (Şekil 2) olarak adlandırılır. Bu gösterimde bir qubit, şu ifadeden elde edilen bir $\vec{r}$ Bloch vektörü ile tanımlanabilir:
> 
> $$\vec{r} = (\cos\phi \sin\theta, \, \sin\phi \sin\theta, \, \cos\theta)$$




Denklem (119)'da $\gamma = \phi = 0$ olan bir kübit durumunu ele alalım. Bu durumda kübitin Bloch küresi gösterimi doğrudan $xz$-düzlemi üzerinde yer alır. Bu gösterimin kullanışlılığı, bir kübit üzerindeki Hadamard dönüşümünün ($H$) etkilerini incelediğimizde anında kendini gösterir. 

Şekil 2'de $(|0\rangle + |1\rangle)/\sqrt{2}$ (yani $|+\rangle$) durumunun, pozitif $x$-ekseni ile kürenin kesişim noktasında bulunduğuna dikkat edin. Buradan yola çıkarak, $H$ kapısının $(|0\rangle + |1\rangle)/\sqrt{2}$ durumu üzerindeki etkisini şu şekilde tarif edebiliriz: $y$-ekseni etrafında $|1\rangle$ durumuna doğru bir dönme ve ardından $xy$-düzleminde bir yansıma. 

Aslında Bloch küresi gösterimi, tüm tek-kübit işlemlerini bu küre üzerindeki dönmeler (rotasyonlar) olarak görmemizi sağlar. Bu derste bunu çok sık kullanmayacak olsak da, tek-kübit uniter işlemlerinin ($U$) Bloch küresi üzerinde nasıl birer dönme olarak ifade edilebildiğini görmek oldukça ilgi çekicidir. $s \in \{x,y,z\}$ eksenlerinden biri etrafında, bir kübit Bloch vektörünü $\theta$ açısı kadar döndüren $R_s(\theta)$ rotasyon matrisi uniter bir işlemdir. Bu matrisler genel olarak şu üstel (exponential) forma sahiptir:

$$R_x(\theta) = e^{-i\theta X / 2}, \quad R_y(\theta) = e^{-i\theta Y / 2} \quad \text{ve} \quad R_z(\theta) = e^{-i\theta Z / 2}$$

Buradaki $X, Y, Z$ ifadeleri Pauli matrisleridir. Bu ders notları için özellikle $z$-ekseni etrafındaki dönme işlemi büyük önem taşımaktadır. Onu daha detaylı bir matris formuyla şu şekilde ifade edebiliriz:

$$R_z(\theta) = e^{-i\theta Z / 2} = \begin{pmatrix} e^{-i\theta/2} & 0 \\ 0 & e^{i\theta/2} \end{pmatrix} = e^{-i\theta/2} \begin{pmatrix} 1 & 0 \\ 0 & e^{i\theta} \end{pmatrix}$$

Herhangi bir rastgele tek-qubit $U$ işlemi, bu dönmeler (rotasyonlar) cinsinden aşağıdaki gibi genel bir formda ifade edilebilir:

$$U = e^{i\alpha} R_z(\beta) R_y(\gamma) R_z(\delta)$$

Buradaki $\alpha, \beta, \gamma$ ve $\delta$ parametreleri birer reel sayıdır.

Son olarak, Bloch kümesi gibi bu kadar temiz ve basit bir geometrik gösterimin **yalnızca tekil kübitler durumu için geçerli olduğunu** unutmamak gerekir: Daha yüksek boyutlu kuantum sistemleri ele alındığında, bir qudit durumunu $d$-boyutlu bir küre cinsinden (bu kadar sezgisel bir biçimde) temsil etmek maalesef mümkün değildir!

## 0.1 Matematiksel Gösterim

Sıkça kullanılan tanımları hatırlayarak başlayalım. $a,b \in \mathbb{R}$ ve $i = \sqrt{-1}$ olmak üzere bir $c = a + ib \in \mathbb{C}$ karmaşık sayısı için, bu sayının karmaşık eşleniğini belirtmek amacıyla $c^* = a - ib$ gösterimini kullanırız. 

Ayrıca kuantum bilişimi boyunca sıklıkla kullanılan özel bir matematiksel gösterime ihtiyacımız olacak. İlk olarak, vektörleri **"bra-ket"** gösterimi olarak bilinen özel bir yöntemle yazacağız. İlk bakışta biraz sıra dışı görünse de, bu vektörlerle yapacağımız birçok işlemi yönetmek için oldukça pratik bir yol sunduğunu göreceksiniz. 

İki örnekle başlayalım. 2 boyutlu bir vektör uzayındaki bir vektörü belirtmek için $|v\rangle \in \mathbb{C}^2$ yazarız. Örneğin:

$$|v\rangle = \begin{pmatrix} 1 \\ 0 \end{pmatrix}$$

Bu vektörün **"bra"** hali, karmaşık eşlenik transpozudur (conjugate transpose). Bizim örneğimiz için bu durum şu şekilde görünür:

$$\langle v| = \left( (|0\rangle)^* \right)^T = \begin{pmatrix} 1^* \\ 0^* \end{pmatrix}^T = \begin{pmatrix} 1 & 0 \end{pmatrix}$$


### Ket ve Bra Gösterimi
$| \cdot \rangle$ ile gösterilen bir **ket**, $\mathbb{C}^d$ karmaşık vektör uzayındaki $d$-boyutlu bir sütun vektörünü (column vector) temsil eder. $\langle \cdot |$ ile gösterilen bir **bra** ise, karşılık gelen ketin karmaşık eşleniğine (complex conjugate) eşit olan $d$-boyutlu bir satır vektörüdür (row vector), yani:

$$\langle \cdot | = (| \cdot \rangle^*)^T$$

Burada $*$ ifadesi her bir elemanın eşleniğini (entry-wise conjugate), $T$ ise transpozu (transpose) ifade eder.

---

Karmaşık sayılarla çalıştığımız için, bu sayıların mutlak değerini de tanımlamamız gerekir.

### Bir Karmaşık Sayının Mutlak Değeri
$x, y \in \mathbb{R}$ sayılarının sırasıyla $z$'nin gerçel (real) ve sanal (imaginary) kısımlarını temsil ettiği, $z = x + iy$ şeklinde ifade edilen bir $z \in \mathbb{C}$ karmaşık sayısını ele alalım. $z$'nin mutlak değeri veya diğer adıyla **modülü**, şu şekilde verilir:

$$|z| = \sqrt{z^* z} = \sqrt{x^2 + y^2}$$

**Örnek:** $z = 1 + i2$ sayısı için mutlak değer $|z| = \sqrt{1^2 + 2^2} = \sqrt{5}$ olarak bulunur.

---

"Bra-ket" gösteriminde iki vektörün **iç çarpımını (inner product)** hesaplamaya sıklıkla ihtiyaç duyacağız. Bu iç çarpım şu şekilde tanımlanır:

### İç Çarpım (Inner Product)
$d$-boyutlu iki vektör verildiğinde:

$$|v_1\rangle = \begin{pmatrix} a_1 \\ \vdots \\ a_d \end{pmatrix} \quad \text{ve} \quad |v_2\rangle = \begin{pmatrix} b_1 \\ \vdots \\ b_d \end{pmatrix}$$

bu iki vektörün iç çarpımı şu şekilde gösterilir ve hesaplanır:

$$\langle v_1 | v_2 \rangle := \langle v_1 || v_2 \rangle = \sum_{i=1}^{d} a_i^* b_i$$


$\mathbb{C}^d$ uzayındaki iki $|v_1\rangle, |v_2\rangle$ vektörünün iç çarpımının genel olarak bir karmaşık sayı (complex number) çıkacağını unutmamak gerekir. İlerleyen bölümlerde, iç çarpımın mutlak değer karesinin ($|\langle v_1|v_2\rangle|^2$) kuantum mekaniğinde ne kadar büyük bir öneme sahip olduğunu göreceğiz. 

Bir örnek olarak, daha önce (2) numaralı eşitlikte verdiğimiz $|v\rangle$ vektörü ile aşağıdaki $|w\rangle$ vektörünün iç çarpımını ele alalım:

$$|w\rangle = \begin{pmatrix} 2 \\ 3 \end{pmatrix}$$

Buradan işlemimiz şu şekilde elde edilir:

$$\langle v|w\rangle = \begin{pmatrix} 1 & 0 \end{pmatrix} \begin{pmatrix} 2 \\ 3 \end{pmatrix} = 1 \cdot 2 + 0 \cdot 3 = 2$$

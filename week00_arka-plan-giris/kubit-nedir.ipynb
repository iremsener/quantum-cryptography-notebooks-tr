# 0.2 Qubit Nedir? (What are qubits?)

Klasik bilgisayarlardaki **bit** kavramına hepimiz sezgisel olarak aşinayız. Peki kuantum bitleri (qubit), klasik bitlerden nasıl ayrılır? 

Bunu görmek için öncelikle klasik bitleri alışılandan biraz daha farklı yazarak başlayalım. Onları sadece `0` ve `1` olarak yazmak yerine, iki farklı vektörle eşleştirelim:

$$0 \rightarrow |0\rangle = \begin{pmatrix} 1 \\ 0 \end{pmatrix} \quad \text{ve} \quad 1 \rightarrow |1\rangle = \begin{pmatrix} 0 \\ 1 \end{pmatrix}$$

> **Not:** Burada Kronecker Delta ($\delta_{ab}$) bağıntısı geçerlidir: Eğer $a \neq b$ ise $\delta_{ab} = 0$, eğer $a = b$ ise $\delta_{ab} = 1$'dir.


Klasik bitlerin doğal kabul ettiğimiz pek çok özelliği vardır; örneğin onları istediğimiz sıklıkta ve serbestçe kopyalayabiliriz (**Kopyalanamama İlkesi / No-Cloning Theorem** nedeniyle kuantum rejiminde bu durum geçerli değildir!). 

Bitlerin fiziksel olarak nasıl uygulanacağını düşündüğümüzde; $|0\rangle$ ve $|1\rangle$ sırasıyla bir atomun **taban (ground)** ve **uyarılmış (excited)** durumlarını temsil edebilir. Bir bit, atomun taban durumunda ($|0\rangle$ için) veya uyarılmış durumunda ($|1\rangle$ için) hazırlanmasıyla kodlanabilir. Bitlerin hayata geçirilmesi için pek çok olası fiziksel yöntem mevcuttur.


### 0.2.1 Tek Bir Kubit (A single qubit)

Durumu vektörler üzerinden düşündüğümüzde, akla şu soru gelir: $\alpha |0\rangle + \beta |1\rangle$ şeklinde herhangi bir vektöre sahip olabilir miyiz? 

İşte kuantum bitlerinin matematiksel tanımı tam olarak budur! Kuantum bitleri sadece kesin bir şekilde "0" veya "1" olmak zorunda değildir; bunun yerine "0" ve "1" durumlarının bir **süperpozisyonunda (superposition / üst üste binme)** bulunabilirler. 

"Quantum bit" ifadesi biraz uzun olduğu için araştırmacılar kısaca **"qubit"** terimini kullanırlar. Bitleri vektör olarak düşündüğümüzde, bir qubit $|v\rangle \in \mathbb{C}^2$ vektörü ile tanımlanabilir. Bu $\mathbb{C}^2$ vektör uzayı, qubitin **durum uzayı (state space)** olarak da bilinir. 

Bir qubit durumuna verilebilecek en temel örneklerden biri $|+\rangle$ (artı) durumudur:

$$|+\rangle = \frac{1}{\sqrt{2}} (|0\rangle + |1\rangle)$$

$\mathbb{C}^2$ uzayındaki her $|v\rangle$ vektörü geçerli bir qubit durumu oluşturur mu? 

Cevap hayır. Geçerli bir qubit olabilmesi için $|v\rangle$ vektörünün **normalize edilmiş (normalized)** olması gerekir. Tıpkı klasik bitlere karşılık gelen $|0\rangle$ ve $|1\rangle$ vektörlerinin de normalize edilmiş olması gibi (bunu kendiniz kontrol edebilirsiniz!). 

Şimdilik bunu bir kural olarak kabul edelim ve bizi aşağıdaki tanıma ulaştırmasına izin verelim:



> 🟦 **Tanım 0.2.1 — Qubit.** Bir qubitin (saf / pure) durumu, 2 boyutlu bir ket vektörü $|\psi\rangle \in \mathbb{C}^2$ ile temsil edilebilir:
> 
> $$|\psi\rangle = \alpha |0\rangle + \beta |1\rangle, \quad \text{burada } \alpha, \beta \in \mathbb{C} \text{ ve } |\alpha|^2 + |\beta|^2 = 1.$$

$\alpha$ ve $\beta$ üzerindeki bu koşul, $|\psi\rangle$ vektörünün normalize edildiği anlamına gelir. Bu karmaşık sayılara (complex numbers) $|\psi\rangle$ durumunun **genlikleri (amplitudes / olasılık genlikleri)** de denir.




> 🟦 **Tanım 0.2.2 — Qudit.** Bir qudit veya $d$-boyutlu bir kuantum sistemi, $d$-boyutlu bir ket vektörü $|\psi\rangle \in \mathbb{C}^d$ ile temsil edilebilir:
> 
> $$|\psi\rangle = \sum_{i=0}^{d-1} \alpha_i |i\rangle, \quad \text{burada } \forall i, \alpha_i \in \mathbb{C} \text{ ve } \sum_{i=0}^{d-1} |\alpha_i|^2 = 1.$$

Buradaki $\alpha_i$ katsayıları üzerindeki koşul, $|\psi\rangle$ vektörünün **uzunluğunun 1 (birim vektör)** olduğu anlamına gelir.


Qubit tanımımızda, klasik bitleri $|0\rangle$ ve $|1\rangle$ vektörleri olarak yazarak işe başlamıştık. Bu iki vektörün **ortonormal (orthonormal)** olduğuna dikkat edin; kuantum notasyonunda bu durum şu şekilde ifade edilir:

$$\langle1|0\rangle = 0 \quad \text{ve} \quad \langle1|1\rangle = \langle0|0\rangle = 1$$

Bu iki vektör $\mathbb{C}^2$ uzayı için bir **taban (baz / basis)** oluşturur. Yani $\mathbb{C}^2$ uzayındaki herhangi bir $|v\rangle$ vektörü, bazı $\alpha, \beta \in \mathbb{C}$ katsayıları için $|v\rangle = \alpha |0\rangle + \beta |1\rangle$ şeklinde yazılabilir. "Klasik" bitlere karşılık gelen bu taban o kadar sık kullanılır ki özel bir adı vardır:

> 🟦 **Tanım 0.2.3 — Standart Taban (Standard Basis).** 2 boyutlu karmaşık vektör uzayı $\mathbb{C}^2$'yi ele alalım. **Standart taban** veya bazen bilinen adıyla **hesapsal taban (computational basis)**, $\mathcal{B} = \{|0\rangle, |1\rangle\}$, bu vektör uzayı için ortonormal bir tabandır ve taban vektörleri şunlardır:
> 
> $$|0\rangle = \begin{pmatrix} 1 \\ 0 \end{pmatrix} \quad \text{ve} \quad |1\rangle = \begin{pmatrix} 0 \\ 1 \end{pmatrix}$$


Elbette $\mathbb{C}^2$ uzayı için tanımlanabilecek pek çok başka taban da mevcuttur. Oldukça sık kullanacağımız bir diğer popüler taban ise aşağıdaki gibi tanımlanan **Hadamard tabanıdır**:

> 🟦 **Tanım 0.2.4 — Hadamard Tabanı (Hadamard Basis).** Hadamard tabanı, şu iki taban elemanından oluşan ortonormal bir $\mathcal{H} = \{|+\rangle, |-\rangle\}$ tabanıdır:
> 
> $$|+\rangle = \frac{1}{\sqrt{2}}(|0\rangle + |1\rangle) = \frac{1}{\sqrt{2}}\begin{pmatrix} 1 \\ 1 \end{pmatrix} \quad \text{ve} \quad |-\rangle = \frac{1}{\sqrt{2}}(|0\rangle - |1\rangle) = \frac{1}{\sqrt{2}}\begin{pmatrix} 1 \\ -1 \end{pmatrix}$$


Şimdi "bra-ket" notasyonunu kullanarak bunun gerçekten ortonormal bir taban olduğunu doğrulayalım. Örnek 0.2.1'de gördüğümüz gibi, $|-\rangle$ zaten normalize edilmiştir. Benzer şekilde $|+\rangle$ durumu için:

$$\langle+|+\rangle = \frac{1}{2}\begin{pmatrix} 1 & 1 \end{pmatrix} \begin{pmatrix} 1 \\ 1 \end{pmatrix} = \frac{1}{2} \cdot 2 = 1 \implies \sqrt{\langle+|+\rangle} = 1$$

olduğundan $|+\rangle$ durumu da normalize edilmiştir (uzunluğu 1'dir). Dahası, bu iki vektörün **iç çarpımına (inner product)** bakarsak:

$$\langle+|-\rangle = \frac{1}{2}\begin{pmatrix} 1 & 1 \end{pmatrix} \begin{pmatrix} 1 \\ -1 \end{pmatrix} = 0$$

sonucunu elde ederiz. Bu da $|+\rangle$ ve $|-\rangle$ vektörlerinin birbirine **dik (orthogonal)** olduğunu gösterir. Hem dik hem de boyları 1 olduğu için bu taban ortonormaldir.


### 0.2.2 Çoklu Kübitler (Multiple qubits)

Klasik olarak iki bitimiz olduğunda, bunları '00', '01', '10' ve '11' şeklinde yazarız. Peki iki kübit'i nasıl yazabiliriz? 

Stratejilerden biri, klasik bitlerin her birini ($x_1, x_2 \in \{0,1\}^2$) yine birer vektörle ilişkilendirmektir. İlk kübit'i $A$, ikinci kübit'i ise $B$ olarak etiketlersek, bit dizilerinden (strings) ortonormal vektörlere şu şekilde bir eşleme (mapping) gerçekleştirebiliriz:

$$0_A 0_B \rightarrow |00\rangle_{AB} = \begin{pmatrix} 1 \\ 0 \\ 0 \\ 0 \end{pmatrix} \quad\quad 0_A 1_B \rightarrow |01\rangle_{AB} = \begin{pmatrix} 0 \\ 1 \\ 0 \\ 0 \end{pmatrix}$$

$$1_A 0_B \rightarrow |10\rangle_{AB} = \begin{pmatrix} 0 \\ 0 \\ 1 \\ 0 \end{pmatrix} \quad\quad 1_A 1_B \rightarrow |11\rangle_{AB} = \begin{pmatrix} 0 \\ 0 \\ 0 \\ 1 \end{pmatrix}$$


Elde edilen bu vektörlerin $d = 2^2 = 4$ boyutuyla $\mathbb{C}^d$ uzayında olduğuna dikkat edin. Buradaki boyut, olası tüm bit dizilerinin sayısına karşılık gelir. 

İki qubitlik bir $|\psi\rangle_{AB} \in \mathbb{C}^4$ durumunu, yine normalizasyon şartını ($\langle\psi|\psi\rangle = 1$) göz önünde bulundurarak bu vektörlerin bir süperpozisyonu şeklinde yazabiliriz. 

Örnek olarak, yukarıdaki tüm standart taban vektörlerinin eşit süperpozisyonundan oluşan bir $|\psi\rangle_{AB}$ durumunu ele alalım:

$$|\psi\rangle_{AB} = \frac{1}{2}|00\rangle_{AB} + \frac{1}{2}|01\rangle_{AB} + \frac{1}{2}|10\rangle_{AB} + \frac{1}{2}|11\rangle_{AB}$$

$$= \frac{1}{2} \begin{pmatrix} 1 \\ 0 \\ 0 \\ 0 \end{pmatrix} + \frac{1}{2} \begin{pmatrix} 0 \\ 1 \\ 0 \\ 0 \end{pmatrix} + \frac{1}{2} \begin{pmatrix} 0 \\ 0 \\ 1 \\ 0 \end{pmatrix} + \frac{1}{2} \begin{pmatrix} 0 \\ 0 \\ 0 \\ 1 \end{pmatrix}$$

$$= \frac{1}{2} \begin{pmatrix} 1 \\ 1 \\ 1 \\ 1 \end{pmatrix}$$


Genliklerin karesinin toplamı $4 \cdot \frac{1}{2^2} = 1$ eder, dolayısıyla $|\psi\rangle$ geçerli bir iki-qubit kuantum durumudur.

Tahmin edebileceğiniz gibi, şimdi $n$ tane qubit'i ele alırken benzer bir mantıkla ilerliyoruz. Çoklu qubitleri ele almak için öncelikle çoklu klasik bitlerin vektör gösterimine bakarız. 

$n$ uzunluğundaki ikili diziler (binary strings) için, her koordinatın $x = x_1, \dots, x_n$ şeklinde bir diziyle etiketlendiği $\mathbb{C}^{2^n}$ vektör uzayını ele alalım. Bu tür dizilerden toplamda $d = 2^n$ adet bulunur; bu nedenle her bir $x$ dizisini $i \in [1, d]$ aralığında farklı bir tam sayı ile etiketleyebiliriz. Ardından, $x$ dizisini, $i$ ile etiketlenen konum hariç her yerde $0$ olan bir $|x\rangle$ vektörü olarak ifade edebiliriz. 

Buna göre, $n$ qubitlik bir kuantum durumu şu şekilde yazılabilir:

$$|\psi\rangle = \sum_{x \in \{0,1\}^n} \alpha_x |x\rangle$$

Burada $\alpha_x \in \mathbb{C}$ ve $\sum_x |\alpha_x|^2 = 1$'dir. $\alpha_x$ sayılarına yine **genlikler (amplitudes)** denir.


Burada $\mathbb{C}^{2^n}$ vektör uzayının boyutunun, bit sayısı olan $n$ ile birlikte **üstel (exponential) olarak arttığını** vurgulamak gerekir. $d = 2^n$ boyutundaki bu $\mathbb{C}^d$ uzayına, $n$ qubitin **durum uzayı (state space)** denir. 

Bu durum, sadece $n$ adet qubitin takibini yapabilmek için bile üstel sayıda $\alpha_x$ parametresine ihtiyacımız olduğu anlamına gelir; bu durum, $n$ adet klasik biti tanımlamak için sadece $n$ adet parametreye ihtiyaç duyan klasik sistemlerle taban tabana zıttır.

Kuanum bitlerini yazmanın tek yolunun bu olup olmadığını merak edebilirsiniz. Sonuçta, $n$ uzunluğundaki dizilerden $\mathbb{C}^d$ uzayındaki vektörlere sadece belirli bir eşleme (mapping) seçmiştik. Dizilerden vektörlere başka bir eşleme seçebilir miydik? 

Cevap evet; her bir dizinin, diğerlerine dik ve boyu 1 olan (**ortonormal**) bir vektöre eşlenmesi şartıyla bu mümkündür. Ancak yukarıdaki eşleme oldukça kullanışlıdır ve kuantum hesaplama dünyasında genel olarak kabul görmüştür. Tek bir qubit durumuna benzer şekilde, $\{|x\rangle \mid x \in \{0,1\}^n\}$ vektör kümesi tarafından verilen bu tabana **standart / hesapsal taban (standard / computational basis)** denir.


> 🟦 **Tanım 0.2.5 — n-qubit için Standart Taban.** $d = 2^n$ olmak üzere, $n$ adet qubitin durum uzayı olan $\mathbb{C}^d$ uzayını ele alalım. Her bir farklı $x \in \{0,1\}^n$ dizisi için, $x$ dizisini $\{1, 2, \dots, d\}$ kümesinden farklı bir $i$ tam sayısı ile ilişkilendirelim. $\mathbb{C}^d$ için standart taban, $\mathcal{S}_n = \{|x\rangle\}_{x \in \{0,1\}^n}$ kümesi tarafından verilen ortonormal bir tabandır.
